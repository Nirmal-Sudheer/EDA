```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel('Data_Train.xlsx')
test_df=pd.read_excel('Test_set.xlsx')
train_df.head()
```

```python
final_df=train_df.append(test_df)
```

```python
#Feature engineering
final_df.info()
```

```python
#converting date_of journey to datetime field
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]

```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
#dropping doj
final_df.drop('Date_of_Journey',axis='columns',inplace=True)
final_df.info()
```

```python
#For arrival time ,we want just the time and not the date that succ it
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1]
final_df.head(1)
```

```python
#converting type
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)

final_df.drop('Arrival_Time',axis='columns',inplace=True)

final_df.info()
```

```python
#For dept time ,we want just the time and not the date that succ it
final_df['Dep_Time']=final_df['Dep_Time'].str.split(' ').str[0]
final_df['Dep_Time'].str.split(' ').str[0]

final_df['Dep_Hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_Min']=final_df['Dep_Time'].str.split(':').str[1]


final_df['Dep_Hour']=final_df['Dep_Hour'].astype(int)
final_df['Dep_Min']=final_df['Dep_Min'].astype(int)

final_df.drop('Dep_Time',axis='columns',inplace=True)
```

```python
final_df.info()
```

```python
#converting total stops to 0-5
final_df['Total_Stops'].unique()

final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})

final_df.head()
```

```python
#drop route
final_df.drop('Route',axis='columns',inplace=True)
```

```python
#converting duration into min
final_df['duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
#above line returns an error cause 2 rows have duration as only 5 min and no h in them.We drop
final_df.drop(6474,axis='rows',inplace=True)
final_df.drop(2660,axis='rows',inplace=True)

final_df['duration_hour']=final_df['duration_hour'].astype(int)
```

```python
final_df['duration_min']=final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
final_df['duration_min'].isnull().sum() #1283 rows are null
#final_df['duration_min']=final_df['duration_min'].astype(int) -> we get an err as o/p as float NaN to int
final_df.drop('Duration',axis='columns',inplace=True)
```

```python
#label encoding for airline
final_df['Airline'].unique()
```

```python
final_df.head(1)
```

```python

from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```

```python
#This can be done instead of the label encoding
final_df=pd.get_dummies(final_df,columns=['Airline','Source','Destination','Additional_Info'],drop_first=True)
final_df.head(2)
```
